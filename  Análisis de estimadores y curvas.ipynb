Paso 2: calcula para cada una de las series los dos valores de la media, la varianza, la cuasivarianza, la desviaci칩n t칤pica y la cuasidesviaci칩n est치ndar. Una vez calculados deben mostrarse en una tabla.


```python
import numpy as np
import pandas as pd


serie1 = [95.5, 96.5, 91.7, 96.1, 94.2, 96.4, 92.8, 96.4, 93.4, 96.9, 92.7, 94.5, 95.2, 96.8, 92.7, 93.1, 99.1, 97.4, 95.8, 92.8]
serie2 = [103.5, 102.8, 101.5, 105.1, 104.2, 102.8, 106.9, 102.9, 103.4, 104.5, 106.9, 102.8, 104.7, 105.4, 106.8, 105.6, 107.4, 109.1, 106.5, 107.2]


def calcular_estadisticas(serie):
    media = np.mean(serie)
    varianza = np.var(serie, ddof=0)
    cuasivarianza = np.var(serie, ddof=1)
    desviacion_tipica = np.std(serie, ddof=0)
    cuasidesviacion_estandar = np.std(serie, ddof=1)
    return [media, varianza, cuasivarianza, desviacion_tipica, cuasidesviacion_estandar]


estadisticas_serie1 = calcular_estadisticas(serie1)
estadisticas_serie2 = calcular_estadisticas(serie2)


estadisticas = pd.DataFrame({
    'Estad칤stica': ['Media', 'Varianza', 'Cuasivarianza', 'Desviaci칩n T칤pica', 'Cuasidesviaci칩n Est치ndar'],
    'Serie 1': estadisticas_serie1,
    'Serie 2': estadisticas_serie2
})


print(estadisticas)
```


Paso 3: a partir de los valores obtenidos de las muestras, estima los valores poblacionales 풮,픢 mediante los estimadores descritos en este tema.


```python
import numpy as np
import pandas as pd


serie1 = [95.5, 96.5, 91.7, 96.1, 94.2, 96.4, 92.8, 96.4, 93.4, 96.9, 92.7, 94.5, 95.2, 96.8, 92.7, 93.1, 99.1, 97.4, 95.8, 92.8]
serie2 = [103.5, 102.8, 101.5, 105.1, 104.2, 102.8, 106.9, 102.9, 103.4, 104.5, 106.9, 102.8, 104.7, 105.4, 106.8, 105.6, 107.4, 109.1, 106.5, 107.2]


def calcular_estadisticas(serie):
    media_muestral = np.mean(serie)
    cuasidesviacion_estandar = np.std(serie, ddof=1)
    return [media_muestral, cuasidesviacion_estandar]

estimadores_serie1 = calcular_estadisticas(serie1)
estimadores_serie2 = calcular_estadisticas(serie2)


estimadores = pd.DataFrame({
    'Estimador': ['Media Poblacional (풮)', 'Desviaci칩n Est치ndar Poblacional (픢)'],
    'Serie 1': estimadores_serie1,
    'Serie 2': estimadores_serie2
})


print(estimadores)
```

Paso 4: representa las curvas correspondientes a la distribuci칩n N(풮,픢) de cada una de las dos series en dos gr치ficos diferentes indicando en cada gr치fica los valores de 풮 y 픢.

Tipifica ambas muestras y comp치ralas con la curva N(0,1) en dos gr치ficos conjuntos.



```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats


serie1 = [95.5, 96.5, 91.7, 96.1, 94.2, 96.4, 92.8, 96.4, 93.4, 96.9, 92.7, 94.5, 95.2, 96.8, 92.7, 93.1, 99.1, 97.4, 95.8, 92.8]
serie2 = [103.5, 102.8, 101.5, 105.1, 104.2, 102.8, 106.9, 102.9, 103.4, 104.5, 106.9, 102.8, 104.7, 105.4, 106.8, 105.6, 107.4, 109.1, 106.5, 107.2]


def calcular_estimadores(serie):
    media = np.mean(serie)
    cuasidesviacion_estandar = np.std(serie, ddof=1)
    return media, cuasidesviacion_estandar


mu1, sigma1 = calcular_estimadores(serie1)
mu2, sigma2 = calcular_estimadores(serie2)


x1 = np.linspace(mu1 - 4*sigma1, mu1 + 4*sigma1, 100)
x2 = np.linspace(mu2 - 4*sigma2, mu2 + 4*sigma2, 100)


y1 = stats.norm.pdf(x1, mu1, sigma1)
y2 = stats.norm.pdf(x2, mu2, sigma2)


plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(x1, y1, label=f'N({mu1:.2f}, {sigma1:.2f})')
plt.title('Distribuci칩n Normal de Serie 1')
plt.xlabel('Valor')
plt.ylabel('Densidad de Probabilidad')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(x2, y2, label=f'N({mu2:.2f}, {sigma2:.2f})')
plt.title('Distribuci칩n Normal de Serie 2')
plt.xlabel('Valor')
plt.ylabel('Densidad de Probabilidad')
plt.legend()

plt.tight_layout()
plt.show()


serie1_tipificada = (serie1 - mu1) / sigma1
serie2_tipificada = (serie2 - mu2) / sigma2


x = np.linspace(-4, 4, 100)
y = stats.norm.pdf(x, 0, 1)


plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(serie1_tipificada, bins=10, density=True, alpha=0.6, color='pink', label='Serie 1 Tipificada')
plt.plot(x, y, label='N(0,1)')
plt.title('Serie 1 Tipificada vs N(0,1)')
plt.xlabel('Valor Tipificado')
plt.ylabel('Densidad de Probabilidad')
plt.legend()

plt.subplot(1, 2, 2)
plt.hist(serie2_tipificada, bins=10, density=True, alpha=0.6, color='purple', label='Serie 2 Tipificada')
plt.plot(x, y, label='N(0,1)')
plt.title('Serie 2 Tipificada vs N(0,1)')
plt.xlabel('Valor Tipificado')
plt.ylabel('Densidad de Probabilidad')
plt.legend()

plt.tight_layout()
plt.show()

```
